```python
data['tags'] = data['tags'].apply(lambda x:x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(data['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem_fun(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
data['tags'] = data['tags'].apply(stem_fun)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
distance = similarity[3]
```

```python
Book_list = sorted(list(enumerate(distance)),reverse=True, key=lambda x:x[1])[1:11]
for i in Book_list:
    print(data['Book-Title'][i[0]])
```

```python
def Book_Recommend(Book):
    Book_index = data[data['Book-Title']==Book].index[0]
    distance = similarity[Book_index]
    Book_list = sorted(list(enumerate(distance)),reverse=True, key=lambda x:x[1])[1:11]
    for i in Book_list:
        print(data['Book-Title'][i[0]])
```
